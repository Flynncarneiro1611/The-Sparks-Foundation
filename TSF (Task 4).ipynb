```python
#Import Libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load the dataset

iris=pd.read_csv("Iris.csv")
```

```python
iris.head()
```

Step 1

```python
X=iris.drop(columns=["species"])
```

```python
y=iris["species"]
```

Step 2

```python
#Splitting the data into training and testing sets

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4,stratify=y)
```

Step 3

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Dtree=DecisionTreeClassifier(criterion='gini')
```

```python
Dtree.fit(X_train,y_train)
```

Step 4

```python
Dtree.score(X_test,y_test)
```

```python
Dtree.predict(X_test)
```

```python
Dtree.predict([[6.3,2.5,5.0,1.9]])
```

```python
#Visualising the Decision Tree
```

```python
names=X.columns
names
```

```python
labels=iris["species"].unique()
labels
```

```python
from sklearn import tree

fig=plt.figure(figsize=(15,8))

fig,ax=plt.subplots(nrows=1,ncols=1,dpi=300,figsize=(4,4))                           #dpi=Dots Per Inches

tree.plot_tree(Dtree,feature_names=names,class_names=labels,filled=True)
plt.show()
```
